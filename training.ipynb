```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from scipy.signal import butter, lfilter

# Define bandpass filter function
def bandpass_filter(signal, low_cutoff=0.5, high_cutoff=50, fs=250, order=2):
    nyquist = 0.5 * fs
    low = low_cutoff / nyquist
    high = high_cutoff / nyquist
    b, a = butter(order, [low, high], btype='band')
    return lfilter(b, a, signal)

# Define function to create binary labels with `1`s only at R-peak positions
def create_binary_labels(rpeak_indices, signal_length, window_size=5):
    """
    Create binary labels with 1s at R-peak positions (expanded by window size) and 0s elsewhere.

    Parameters:
    - rpeak_indices: List of indices where R-peaks are located
    - signal_length: The length of the signal (e.g., 2500 data points per signal)
    - window_size: The size of the window to expand around each R-peak. Default is 5.
    
    Returns:
    - labels: A binary array of the same length as the signal, with 1s at R-peak positions and 0s elsewhere.
    """
    labels = np.zeros(signal_length)  # Initialize the labels as zeros
    half_window = window_size // 2    # Half window to center around R-peak

    for index in rpeak_indices:
        # Ensure we stay within the signal boundaries when applying the window
        start = max(0, index - half_window)
        end = min(signal_length, index + half_window + 1)
        labels[start:end] = 1  # Set the region around the R-peak to 1

    return labels

# Generate binary labels for each ECG signal
def generate_binary_labels(df):
    """
    Generate binary labels for all ECG signals in the dataframe.
    
    Parameters:
    - df: DataFrame containing 'ECG Signal' and 'R Indices'
    
    Returns:
    - labels_list: A list of binary label arrays, one for each ECG signal.
    """
    labels_list = []
    
    for _, row in df.iterrows():
        rpeak_indices = row['R Indices']         # Get the R-peak indices for this signal
        signal_length = len(row['ECG Signal'])  
        labels = create_binary_labels(rpeak_indices, signal_length)
        labels_list.append(labels)               # Add the binary labels to the list
    
    return np.array(labels_list)

# Prepare filtered signals by applying the bandpass filter to combined ECG and noise signals
def prepare_data(df):
    filtered_signals = []
    for _, row in df.iterrows():
        ecg_signal = row['ECG Signal']
        noise_signal = row['Noise Signal']
        combined_signal = ecg_signal+noise_signal 
        filtered_signal = bandpass_filter(combined_signal)
        filtered_signals.append(filtered_signal)
    return np.array(filtered_signals)

def scale_data(filtered_signals):
    scaler = MinMaxScaler(feature_range=(-1, 1))
    signals_reshaped = filtered_signals.reshape(-1, 1)
    scaled_signals = scaler.fit_transform(signals_reshaped)
    scaled_signals = scaled_signals.reshape(filtered_signals.shape)
    return scaled_signals

# Keep only the required fields from the dataframes
afib_df = afib_df[['ECG Signal', 'Noise Signal', 'R Indices']]
sinus_df = sinus_df[['ECG Signal', 'Noise Signal', 'R Indices']]

# Combine the datasets
combined_df = pd.concat([afib_df, sinus_df], ignore_index=True)
combined_df = combined_df.sample(frac=1, random_state=39).reset_index(drop=True)

# Prepare filtered signals (apply bandpass filter, etc.)
filtered_signals = prepare_data(combined_df)

# Generate binary labels for R-peak detection
labels = generate_binary_labels(combined_df)

# Normalize/scale the filtered signals
training = scale_data(filtered_signals)

print(f"Shape of training signals: {training.shape}")
print(f"Shape of labels: {labels.shape}")
```

### Dataset contains highly imbalanced classes of peaks and non peaks. That's why I tried multiple strategies to cater that

```python

# Count the number of 0's and 1's
class_0_count = np.sum(labels == 0)  # Majority class count (assuming 0 is majority)
class_1_count = np.sum(labels == 1)  # Minority class count (assuming 1 is minority)

# Identify the majority and minority class counts
minority_class_count = min(class_0_count, class_1_count)
majority_class_count = max(class_0_count, class_1_count)

# Calculate the imbalance ratio
imbalance_ratio = minority_class_count / majority_class_count

# Print the results
print(f"Minority Class Count: {minority_class_count}")
print(f"Majority Class Count: {majority_class_count}")
print(f"Class Imbalance Ratio: {imbalance_ratio:.4f}")
```

## Model Creation 
##### Bidrectional LSTM 



```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Bidirectional, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(training, labels, test_size=0.2, random_state=35, shuffle= True)

# Reshape the input for the LSTM layer
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1)) 
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1)) 

# Count the number of R-peak and non-R-peak labels in the training set
r_peak_count = np.sum(y_train)  # Total number of 1s (R-peaks)
non_r_peak_count = np.prod(y_train.shape) - r_peak_count  # Total number of 0s (non-R-peaks)

# Calculate class weights (higher weight for R-peaks)
weight_for_r_peak = (1 / r_peak_count) * (y_train.shape[0] / 2.0)
weight_for_non_r_peak = (1 / non_r_peak_count) * (y_train.shape[0] / 2.0)

class_weights = {0: weight_for_non_r_peak, 1: weight_for_r_peak}

# Define the BiLSTM model for R-peak detection
model = Sequential()

# Add a Bidirectional LSTM layer
model.add(Bidirectional(LSTM(64, return_sequences=True), input_shape=(2500, 1)))
#model.add(Dropout(0.3))

# Add another Bidirectional LSTM layer
model.add(Bidirectional(LSTM(64, return_sequences=True)))
#model.add(Dropout(0.3))

# Output layer for binary classification per time step
model.add(Dense(1, activation='sigmoid'))


# Using focal loss to cater class imbalance
def focal_loss(gamma=2., alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        eps = tf.keras.backend.epsilon()
        y_pred = tf.keras.backend.clip(y_pred, eps, 1. - eps)
        pt = tf.where(tf.equal(y_true, 1), y_pred, 1 - y_pred)
        CE = -tf.keras.backend.log(pt)
        FL = tf.keras.backend.pow(1 - pt, gamma) * CE
        return tf.keras.backend.mean(alpha * FL)
    return focal_loss_fixed

# Compile the model with focal loss
model.compile(optimizer=Adam(learning_rate=1e-4), loss=focal_loss(gamma=2., alpha=0.25), metrics=['accuracy'])

# Add a learning rate scheduler to reduce learning rate when validation loss plateaus
lr_schedule = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.01, patience=5, min_lr=1e-6)

# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test), class_weight=class_weights, callbacks=[lr_schedule])
# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"Test Accuracy: {test_acc:.4f}")

# Predict R-peaks on the test set
y_pred = (model.predict(X_test) > 0.5).astype(int) 

# Flatten the predictions and true labels for classification report and confusion matrix
print("Classification Report:\n", classification_report(y_test.flatten(), y_pred.flatten()))
print("Confusion Matrix:\n", confusion_matrix(y_test.flatten(), y_pred.flatten()))
model.save("training.keras")
```

```python
from scipy.ndimage import binary_dilation
from sklearn.metrics import classification_report, confusion_matrix

# Evaluate the model

test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"Test Accuracy: {test_acc:.4f}")

# Predict R-peaks on the test set
y_pred = model.predict(X_test)

# Apply threshold
y_pred_binary = (y_pred > 0.5).astype(int)  # Ensure y_pred is in binary format

# Check the shape after thresholding
print(f"Shape of y_pred_binary: {y_pred_binary.shape}")

y_pred_binary_reshaped = y_pred_binary.reshape(y_pred_binary.shape[0], y_pred_binary.shape[1])

print(f"Shape of y_pred_binary_reshaped: {y_pred_binary_reshaped.shape}")

# Apply binary dilation
# Use a 2D structure for 2D input
structure = np.ones((1, 5), dtype=bool)  # 1x5 structure for dilation

# Perform binary dilation
y_pred_processed = binary_dilation(y_pred_binary_reshaped, structure=structure)

print(f"Shape of y_pred_processed: {y_pred_processed.shape}")

print("Classification Report after post-processing:\n", classification_report(y_test.flatten(), y_pred_processed.flatten()))
print("Confusion Matrix after post-processing:\n", confusion_matrix(y_test.flatten(), y_pred_processed.flatten()))
```
